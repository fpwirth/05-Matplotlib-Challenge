Load Datasets and Cleanup Data

```python
#Load Dependencies
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Select data files and import
mousefile = os.path.join('data', 'mouse_drug_data.csv')
trialfile = os.path.join('data', 'clinicaltrial_data.csv')
mousedata = pd.read_csv(mousefile)
trialdata = pd.read_csv(trialfile)

#Merge data files
tmouse = pd.merge(trialdata, mousedata, on=['Mouse ID'], how='left')
```

```python
#Check merged file for proper merge
tmouse.shape
```

```python
#Check clinical trail data file size
trialdata.shape
```

```python
#Check mouse drug data file size
mousedata.shape
```

```python
#Find duplicates in mouse drug data file
mousedups = mousedata[mousedata.duplicated(['Mouse ID'], keep=False)]
mousedups
```

```python
#Drop Mouse IDs with drugs not involved in analysis and check for null values (ie. make sure drug for all mouse ids)
mouse = tmouse.loc[tmouse['Drug'].isin(['Capomulin','Infubinol','Ketapril','Placebo'])]
mouse.isnull().values.any()
```

```python
# Display the data table for preview
mouse
```

Tumor Response to Treatment

```python
# Compute the mean for tumor volume grouped by drug and timepoint 
tumorvolume = mouse.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean().unstack(0)
tumorvolume.reset_index(inplace=True)

# Compute the sem for tumor volume grouped by drug and timepoint
tumorvolumesem = mouse.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem().unstack(0)
tumorvolumesem.reset_index(inplace=True)
```

```python
# Generate the tumor volume plot (with error bars)
plt.errorbar(tumorvolume['Timepoint'],tumorvolume['Capomulin'],tumorvolumesem['Capomulin'],marker='o',c='red',label='Capomulin')
plt.errorbar(tumorvolume['Timepoint'],tumorvolume['Infubinol'],tumorvolumesem['Infubinol'],marker='^',c='blue',label='Infubinol')
plt.errorbar(tumorvolume['Timepoint'],tumorvolume['Ketapril'],tumorvolumesem['Ketapril'],marker='s',c='green',label='Ketapril')
plt.errorbar(tumorvolume['Timepoint'],tumorvolume['Placebo'],tumorvolumesem['Placebo'],marker='d',c='black',label='Placebo')
plt.legend(loc='best')
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.xlim=len(tumorvolume['Timepoint'])

plt.savefig('images/treatment.jpg')
```

Metastatic Response to Treatment

```python
# Compute the mean Metastatic sites grouped by drug and timepoint
msites = mouse.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean().unstack(0)
msites.reset_index(inplace=True)

# Compute the sem Metastatic sites grouped by drug and timepoint
msitessem = mouse.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem().unstack(0)
msitessem.reset_index(inplace=True)
```

```python
# Generate the metastatic sites plot (with error bars)
plt.errorbar(msites['Timepoint'],msites['Capomulin'],msitessem['Capomulin'],marker='o',c='red',label='Capomulin')
plt.errorbar(msites['Timepoint'],msites['Infubinol'],msitessem['Infubinol'],marker='^',c='blue',label='Infubinol')
plt.errorbar(msites['Timepoint'],msites['Ketapril'],msitessem['Ketapril'],marker='s',c='green',label='Ketapril')
plt.errorbar(msites['Timepoint'],msites['Placebo'],msitessem['Placebo'],marker='d',c='black',label='Placebo')
plt.legend(loc='best')
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Metastatic Sites')
plt.xlim=len(msites['Timepoint'])

plt.savefig('images/spread.jpg')
```

Survival Rates

```python
# Compute the percent change of tumor volume for each drug
survival = mouse.groupby(['Drug', 'Timepoint'])['Timepoint'].count().unstack(0)
survival.reset_index(inplace=True)

survival['pCapomulin'] = (survival['Capomulin']/survival['Capomulin'][0])*100
survival['pInfubinol'] = (survival['Infubinol']/survival['Infubinol'][0])*100
survival['pKetapril'] = (survival['Ketapril']/survival['Ketapril'][0])*100
survival['pPlacebo'] = (survival['Placebo']/survival['Placebo'][0])*100
```

```python
# Generate the survival plot
plt.plot(survival['Timepoint'],survival['pCapomulin'],marker='o',c='red',label='Capomulin')
plt.plot(survival['Timepoint'],survival['pInfubinol'],marker='^',c='blue',label='Infubinol')
plt.plot(survival['Timepoint'],survival['pKetapril'],marker='s',c='green',label='Ketapril')
plt.plot(survival['Timepoint'],survival['pPlacebo'],marker='d',c='black',label='Placebo')
plt.legend(loc='best')
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.xlim=len(survival['Timepoint'])

plt.savefig('images/survival.jpg')
```

Summary Bar Graph

```python
change=pd.DataFrame({'drug':['Capomulin','Infubinol','Ketapril','Placebo']})
capomulin=((tumorvolume['Capomulin'][tumorvolume.index[-1]]-tumorvolume['Capomulin'][0])/tumorvolume['Capomulin'][0])*100
infubinol=((tumorvolume['Infubinol'][tumorvolume.index[-1]]-tumorvolume['Infubinol'][0])/tumorvolume['Infubinol'][0])*100
ketapril=((tumorvolume['Ketapril'][tumorvolume.index[-1]]-tumorvolume['Ketapril'][0])/tumorvolume['Ketapril'][0])*100
placebo=((tumorvolume['Placebo'][tumorvolume.index[-1]]-tumorvolume['Placebo'][0])/tumorvolume['Placebo'][0])*100
change['change']=[capomulin,infubinol,ketapril,placebo]
```

```python
# Generate the change in tumor volume bar chart
changeplt=plt.bar(change['drug'],change['change'],color=(change['change']<0).map({True: 'g', False:'r'}))
plt.title('Tumor Change Over 45 Day Treatment')
plt.xlabel('Drug Administered')
plt.ylabel('% Tumor Volume Change')
plt.xlim=len(change['drug'])
plt.ylim(-30,70)
#add horizontal line at zero
plt.axhline(y=0, color = 'black')

count=0
for c in change['change']:
    if c < 0:
        y_coor=-5
    else:
        y_coor=2
    plt.text(count,y_coor,str(round(c,1))+'%',ha='center',c='w')
    count+=1

plt.savefig('images/change.jpg')
```
